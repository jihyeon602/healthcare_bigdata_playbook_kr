```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('stackPerDevice_datetime.csv')
df1.head()
```

```python
df1.groupby(['deviceId','date', 'status'])['cnt']
```

```python
df1.groupby(['deviceId','date', 'status'])['cnt'].count()
```

```python
df1.groupby(['deviceId','date', 'status'])[['cnt']].sum()
```

```python
df1.groupby(['deviceId','date', 'status'])['cnt'].sum()
```

```python
df1.groupby(['deviceId','date', 'status'])['cnt'].sum().unstack()
```

```python
pt = df1.pivot_table('cnt', index=['deviceId','date'], columns='status')
pt
```

```python
pt = df1.pivot_table('cnt', index=['deviceId','date'], columns='status', aggfunc='sum')
pt
```

```python
#melt
df1.melt(id_vars=['deviceId','date'], value_vars=['status'])
```

```python
pt = df1.pivot_table(index=['deviceId','date'], columns='status', aggfunc={'cnt': sum, 'status': 'count'})
```
