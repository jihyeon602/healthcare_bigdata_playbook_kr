```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('grayscale')

# plt.figure(figsize=(10, 10))

# 데이터 로드
df = pd.read_csv('data/kr_population.csv', index_col=0, parse_dates=True)
df.head()
```

```python

# 년도별 출생자수
df['birth'].plot(label='Num of prople born in S.korea', linestyle='-', figsize=(8,6))

# 년도별 출생자수 10년 이동 평균하여, 선을 부드럽게 그리기
df['birth'].rolling(window=10).mean().plot(label='Rolling Mean (window=10)', linestyle=":")

# # 누적 평균을 그리기
df['birth'].expanding().mean().plot(label='Expanding', linestyle="--")

# plt.rcParams["figure.figsize"] = [10, 10]
# plt.rcParams["figure.figsize"] = [20, 10]
plt.legend()
```

```python
# fillna로 결측치 채우기
# 년도별 출산율
df['birthrate'].fillna(method='ffill').plot(label='Population growth Rate', linestyle='-', figsize=(8,6))

# 년도별 출산율 10년 이동 평균하여, 선을 부드럽게 그리기
df['birthrate'].fillna(method='ffill').rolling(window=10).mean().plot(label='Rolling Mean (window=10)', linestyle=":")

# 누적 평균을 그리기
df['birthrate'].fillna(method='ffill').expanding().mean().plot(label='Expanding', linestyle="--")

plt.axhline(y=2.0, xmin=0.6, xmax=1.0, color='lightgray', linestyle='--', linewidth=2)
plt.axhline(y=1.0, xmin=0.6, xmax=1.0, color='lightgray', linestyle='-', linewidth=2)

plt.legend()
plt.show()
```

```python
# 10년간 누적 출생자수 계산
df['birth'].rolling(window=10).sum().plot(label='Rolling Sum (window=10)', linestyle=":", figsize=(8,6))

# 누적 출생자수를 그리기
df['birth'].expanding().sum().plot(label='Expanding', linestyle="-.")

plt.legend()
```
