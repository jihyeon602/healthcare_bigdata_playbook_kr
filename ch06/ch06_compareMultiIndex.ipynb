```python
import pandas as pd
import numpy as np
import timeit

# Index로 구성된 데이터프레임
def create_df_index():
    df = pd.DataFrame({
        "A": np.random.randint(0, 100, 1000000),
        "B": np.random.randint(0, 100, 1000000),
        "C": np.random.randint(0, 100, 1000000)
    })
    return df

# MultiIndex로 구성된 데이터프레임
def create_df_multiindex():
    df = pd.DataFrame({
        "A": np.random.randint(0, 100, 1000000),
        "B": np.random.randint(0, 100, 1000000),
        "C": np.random.randint(0, 100, 1000000)
    })
    df = df.set_index(["A", "B"])
    return df

index_time = timeit.timeit(create_df_index, number=10)
multiindex_time = timeit.timeit(create_df_multiindex, number=10)

print("Index로 구성된 데이터프레임의 실행 시간: ", index_time)
print("MultiIndex로 구성된 데이터프레임의 실행 시간: ", multiindex_time)

if index_time < multiindex_time:
    print("Index로 구성된 데이터프레임의 성능이 더 좋습니다.")
else:
    print("MultiIndex로 구성된 데이터프레임의 성능이 더 좋습니다.")
```
